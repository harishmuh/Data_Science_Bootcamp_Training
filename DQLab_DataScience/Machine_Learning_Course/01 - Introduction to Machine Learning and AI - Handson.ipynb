## Dictionary

```python
daftar_belanja = {
    'baju':75000,
    'celana':125000,
    'topi':25000
}

belanja = ['baju', 'celana', 'topi', 'baju']
```

```python
harga = [daftar_belanja[x] for x in belanja]
harga
```

```python
total_belanja = sum(harga)
total_belanja
```

# Method and Function

```python
def luas_keliling(panjang, lebar):
    luas = panjang * lebar
    keliling = 2*(panjang+lebar)
    print(f"Luas lingkaran : {luas}")
    print(f"Keliling lingkaran : {keliling}")
```

```python
luas_keliling(10, 3)
```
